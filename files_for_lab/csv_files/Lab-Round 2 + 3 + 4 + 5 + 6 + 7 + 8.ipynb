```python
#Differentiating the categorical and numerical features as they're going to be treated differently
X2_num = X2.select_dtypes(include = np.number)
X2_cat = X2.select_dtypes(include = object)
```

```python
#Dropping the following columns because the values are not essential when creating the model, and there are too many unique values 
X2_cat = X2_cat.drop(['customer'],axis =1)
```

```python
X2_cat.dtypes
```

```python
X2_cat.shape
```

```python
# Exercise 1. : OneHot Encoding
#Step 1. 
#We need to train the encoder with a command to create dummy variables
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(drop='first').fit(X2_cat)

#2. Show the new categories
print(encoder.categories_)

# 3. Change the variable into an array
encoded = encoder.transform(X2_cat).toarray()
print(encoded)

# 4. Create cols
cols=[]
for row in encoder.categories_:
    for colname in row:
        cols.append(colname)
cols

# 5. Add names to the features again
cols = encoder.get_feature_names(input_features=X2_cat.columns)
cols

# 6. Let's see if it worked
onehot_encoded = pd.DataFrame(encoded, columns=cols)
onehot_encoded.head(2)
```

```python
#Using the MinMax Scaler, used this to normalize the data for X2_num (process was done in Round 5, although repeating it and still getting same data.shape(10910,9))
```

```python
from sklearn.preprocessing import MinMaxScaler
transformer = MinMaxScaler().fit(X2_num)
#transformer contains the data to do the transformation
x_normalized = transformer.transform(X2_num)
#normalized will show the scaled values
print(x_normalized.shape)
x_normalized
normalized_data= pd.DataFrame(x_normalized, columns=X2_num.columns)
```

```python
#Exercise 2.: Concat DataFrames
```

```python
X = pd.concat([normalized_data, onehot_encoded], axis=1)
```
